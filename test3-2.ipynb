```python
#test3-2 2023.3.13
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

X=[10, 15, 20, 25, 30, 35]
Y=[136, 140, 155, 160, 157, 175]
X1=np.array(X).reshape(-1,1)
Y1=np.array(Y).reshape(-1,1)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X1,Y1)
print(model.coef_)
print(model.intercept_)

import matplotlib.pyplot as plt

Y2= model.predict(X1)
plt.scatter(X1,Y1,c='g')
plt.plot(X1,Y2,c= 'b')

X3= PolynomialFeatures(degree=3).fit_transform(X1)

print(X3)

from sklearn.linear_model import LinearRegression

model3=LinearRegression()
model3.fit(X3,Y1)
print(model3.coef_)

#Y3=model3.predict(X3) 
#import matplotlib.pyplot as plt
#plt.plot(X1,Y3,c='b') # not a curve 

X2= np.linspace(10,35,100)
X4=np.array(X2).reshape(-1,1)

Y4=model3.predict(PolynomialFeatures(degree=3).fit_transform(X4))
            
plt.plot(X4, Y4, c='r')
```
