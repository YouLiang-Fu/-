```python
import numpy as np
samples=np.loadtxt("linear.txt")

import matplotlib.pyplot as plt
plt.scatter(samples[:,0],samples[:,1],c=samples[:,2])

from sklearn.linear_model import LogisticRegression
X=samples[:,0:2]
Y=samples[:,2]
logreg=LogisticRegression()
logreg.fit(X,Y)
print(logreg.coef_)             # intercept_+ coef_[0,0]x+ coef_[0,1]y =0 
print(logreg.intercept_)     
coef_= logreg.coef_             # intercept_+ coef_[0,0]x+ coef_[0,1]y =0 
intercept_= logreg.intercept_
X1= samples[:,0] 
# y= -coef_[0,0]/coef_[0,1]x- intercept_/coef_[0,1]
Y1= -coef_[0,0]/coef_[0,1]*X1- intercept_/coef_[0,1] 
print(-coef_[0,0]/coef_[0,1])
print(- intercept_/coef_[0,1])
plt.plot(X1,Y1,c='b')
plt.show()
```
