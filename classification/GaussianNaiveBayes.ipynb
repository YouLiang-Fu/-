```python
import numpy as np
samples=np.loadtxt("linear.txt")

import matplotlib.pyplot as plt 
plt.scatter(samples[:,0],samples[:,1],c=samples[:,2])

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
X=samples[:,0:2]
y=samples[:,2]
gnb.fit(X, y)
gnb.score(X,y)

```

```python
y_pred = gnb.predict(X)
plt.scatter(samples[:,0],samples[:,1],c=y_pred)
```
